```python
df = df.rename(columns={
    'bugID': 'Bug ID',
    'ct': 'Creation Time',
    'sd': 'Summary Description',
    'dt': 'Date Target',
    'cl': 'Category Label',
    'pd': 'Priority Description',
    'co': 'Component',
    'rp': 'Reporter',
    'os': 'Operating System',
    'bs': 'Build System',
    'rs': 'Resolution Status',
    'pr': 'Priority',
    'bsr': 'Build System Revision',
    're': 'Regression',
    'at': 'Assigned To'
})
```

```python
unique_categories = df["Priority"].unique()
print(unique_categories)
```

```python
df["Priority"] = df["Priority"].replace("--", "P0")
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
from warnings import filterwarnings
stop_words = stopwords.words('english')
filterwarnings('ignore')

def remove_stopwords(text):
    words = text.lower().split()
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)

df['Summary Description'] = df['Summary Description'].apply(remove_stopwords)
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
summary_vectors = tfidf.fit_transform(df['Summary Description'])
```
